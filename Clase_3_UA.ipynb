```python
# prompt: crea una tabla de frecuencia completa, el histograma con su poligono de frecuencia y el diagrama de caja  con la variable peso

# Calcular la tabla de frecuencia
peso = df['PESO'].dropna() # Eliminar valores nulos en la variable PESO
tabla_frecuencia = pd.DataFrame(peso.value_counts()).reset_index()
tabla_frecuencia.columns = ['Peso', 'Frecuencia Absoluta']
tabla_frecuencia['Frecuencia Relativa'] = tabla_frecuencia['Frecuencia Absoluta'] / len(peso)
tabla_frecuencia['Frecuencia Acumulada'] = tabla_frecuencia['Frecuencia Absoluta'].cumsum()
tabla_frecuencia['Frecuencia Relativa Acumulada'] = tabla_frecuencia['Frecuencia Relativa'].cumsum()
print("Tabla de Frecuencia para la variable PESO:")
print(tabla_frecuencia)

# Crear el histograma y el polígono de frecuencia
plt.figure(figsize=(10, 6))
plt.hist(peso, bins='auto', density=True, alpha=0.7, label='Histograma') # 'auto' para determinar el número de bins automáticamente
sns.kdeplot(peso, color='red', label='Polígono de Frecuencia (Estimación de Densidad)')
plt.title('Histograma y Polígono de Frecuencia de la variable PESO')
plt.xlabel('Peso')
plt.ylabel('Densidad')
plt.legend()
plt.grid(axis='y', alpha=0.5)
plt.show()

# Crear el diagrama de caja
plt.figure(figsize=(8, 6))
sns.boxplot(x=peso)
plt.title('Diagrama de Caja de la variable PESO')
plt.xlabel('Peso')
plt.grid(axis='x', alpha=0.5)
plt.show()
```

```python
# prompt: EN EL DIAGRAMA DE CAJA DE DATOS ANTERIOR NECESITO QUE LE DES UN ORDEN ASCENTE A LA VARIABLE ESTRATO.

# Assuming your data is in a pandas DataFrame called 'df'
# and the columns are named 'PARCIAL 2', 'SEXO', and 'ESTRATO'

# Convert the 'ESTRATO' column to string type before getting unique values
df['ESTRATO'] = df['ESTRATO'].astype(str)

# Convert the 'ESTRATO' column to a categorical type with a specific order
estrato_order = sorted(df['ESTRATO'].unique())
df['ESTRATO'] = pd.Categorical(df['ESTRATO'], categories=estrato_order, ordered=True)


plt.figure(figsize=(10, 6))
sns.boxplot(x='ESTRATO', y='PARCIAL 2', hue='SEXO', data=df)
plt.title('Diagrama de Caja de PARCIAL 2 por ESTRATO y SEXO (Ordenado)')
plt.xlabel('ESTRATO')
plt.ylabel('PARCIAL 2')
plt.show()
```
